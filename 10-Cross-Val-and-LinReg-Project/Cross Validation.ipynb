# cross_validate

```python
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.model_selection import cross_validate
```

```python
model = Ridge(alpha=100)
```

```python
scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'], cv=10)
```

```python
scores = pd.DataFrame(scores)
```

```python
scores # Scores for each fold with fitting time and error metrics
# More info than cross_val_score
```

```python
# get means out
scores.mean()
```

```python
model = Ridge(alpha=1)
```

```python
scores = cross_validate(model, X_train, y_train, scoring=['neg_mean_squared_error', 'neg_mean_absolute_error'], cv=10)
```

```python
scores = pd.DataFrame(scores)
```

```python
# get means out
scores.mean() # Much better performance.
```

```python
# Fit on holdout data
model.fit(X_train, y_train)
```

```python
# Predict off of x test data
y_final_pred = model.predict(X_test)
```

```python
# Final reported score
# Get the predictions and pit them against the y test holdout set
mean_squared_error(y_test, y_final_pred)
```

# Grid Search

- Way of training and validating a model on every possible combination of multiple hyperparam options

```python
# Via GridSearchCV class
## CREATE X and y
X = df.drop('sales',axis=1)
y = df['sales']

# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# SCALE DATA
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
help(ElasticNet)
```

```python
# What is the optimal alpha and l1_ration value...
```

```python
base_elasticnet_model = ElasticNet()
```

```python
# Create grid of params and values
param_grid = {
    'alpha': [0.1, 1, 5, 10, 50, 100],
    'l1_ratio': [.1, .5, .7, .95, .99, 1]
}
```

```python
# Perform the gird search
from sklearn.model_selection import GridSearchCV
```

```python
# Paying for simplicity with computational time.
grid_model = GridSearchCV(
    estimator=base_elasticnet_model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=5,
    verbose=2 # Can play around with how verbose you want it to be.
)
```

```python
grid_model.fit(X_train, y_train) # Another form of cross validation but tests different param choices!
```

```python
grid_model.best_estimator_ # Gives me back the best ML model to use!!!! what!!!!!
```

```python
grid_model.best_params_
```

```python
# See results for all params!
pd.DataFrame(grid_model.cv_results_)
```

```python
# We can call predict off the gird_model's best estimator!!
y_pred = grid_model.predict(X_test)
```

```python
# Final reportable results...
mean_squared_error(y_test, y_pred)
```
