# Classification of text

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/airline_tweets.csv')
```

```python
df
```

```python
sns.countplot(data=df, x='airline_sentiment')
```

```python
sns.countplot(data=df, x='negativereason')
plt.xticks(rotation=90)
```

```python
# Get out our y and X
data = df[['airline_sentiment', 'text']]
```

```python
data
```

```python
X = data['text']
y = data['airline_sentiment']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectoriser = TfidfVectorizer(stop_words='english')
```

```python
tfidf_vectoriser.fit(X_train)
```

```python
# Get the tfidf versions
X_train_tfidf = tfidf_vectoriser.transform(X_train)
X_test_tfidf = tfidf_vectoriser.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.svm import SVC, LinearSVC
```

```python
rbf_svc = SVC()
rbf_svc.fit(X_train_tfidf, y_train)
```

```python
linear_svc = LinearSVC()
linear_svc.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
def report(model):
    preds = model.predict(X_test_tfidf)
    print(classification_report(y_test, preds))
    plot_confusion_matrix(model, X_test_tfidf, y_test)
```

```python
report(nb)
```

```python
report(log_model)
```

```python
report(rbf_svc)
```

```python
report(linear_svc)
```

```python
from sklearn.pipeline import Pipeline
```

```python
# Technically model deployment
pipe = Pipeline([('tfidf', TfidfVectorizer()), ('svc', LinearSVC())])
```

```python
pipe.fit(X, y)
```
