```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATA/Ames_NO_Missing_Data.csv")
```

```python
df.head()
```

```python
df['MS SubClass'] = df['MS SubClass'].apply(str)
```

```python
direction = pd.Series(['Up', 'Up', 'Down'])
direction
```

```python
# Column of string features to ONE HOT ENCODING!
pd.get_dummies(direction, drop_first=True)
```

```python
# Check what columns are string (pandas thinks string = object)
df.select_dtypes(include='object')
```

```python
# Convert to one hot
my_object_df = df.select_dtypes(include='object')
my_numeric_df = df.select_dtypes(exclude='object')
```

```python
df_objects_dummies = pd.get_dummies(my_object_df, drop_first=True)
```

```python
# One hot encoded all string values.
df_objects_dummies.head()
```

```python
# Combine the DFs
finaldf = pd.concat([my_numeric_df, df_objects_dummies], axis=1)
```

```python
finaldf
```

```python
# See what featurs correlate to the SalePrice feature
finaldf.corr()['SalePrice'].sort_values()
```
