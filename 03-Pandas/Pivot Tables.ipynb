# Pivot Tables

- a DataFrame with repeated values can be pivoted for reorganization and clarity.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
df
```

```python
# How many licenses of each product type did google purchase.
# We can try reformatting (choose 3 cols we want to work with)

licenses = df[['Company', 'Product', 'Licenses']]
licenses
```

```python
# Index will be repeating values.
pd.pivot(licenses, index='Company', columns='Product', values='Licenses')
```

```python
# what is the total sum for each company.
# You can do this by group by probs easier.
pd.pivot_table(df, index='Company', aggfunc='sum', values=['Licenses', 'Sale Price'])
```

```python
# or...
df.groupby('Company').sum()
```

```python
# Groupd by acc manager, and contact and get total sale price per contact.
pd.pivot_table(df, index=['Account Manager', 'Contact'], values=['Sale Price'], aggfunc='sum')
```
