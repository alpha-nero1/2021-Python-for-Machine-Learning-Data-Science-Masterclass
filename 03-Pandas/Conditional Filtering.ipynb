# Conditional Filtering

- We may want to filter by condition not index.
- Allows us to select rows from column condition.
- Organising data really comes into play here.
- Each column = feature/attribute
- Each row = instance of the data.
- We can always transpose rows and columns

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
# Filter by single condition
bool_series = df['total_bill'] > 40

# The condition was broadcast across the series returned
bool_series
```

```python
# Only get us rows where boolean series index match was true
# This applies a filter.
df[bool_series]
```

```python
# Typically we would just write
df[df['total_bill'] > 40]
```

```python
# Filter by male
df[df['sex'] == 'Male']
```

```python
# Filter by size
df[df['size'] > 3]
```

## Filter by multiple conditions

```python
# Can use AND & operator for AND
# Can use OR | for OR
1 == 1 or 2 == 2
```

```python
df['total_bill'] > 30

df['sex'] == 'Male'
```

```python
# We MUST use ampersand or pipe for multiple conditions!
# These symbols let us chain the boolean series together...
both = df[((df['total_bill'] > 30) & ((df['sex'] == 'Male')))]
both
```

```python
# We MUST use ampersand or pipe for multiple conditions!
either = df[((df['total_bill'] > 30) | ((df['sex'] == 'Male')))]
either
```

```python
# Filter by list of options...
df
```

```python
# For nominal attributes.
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```

```python
# Concatinated/better example, use isin method (isin = Series function)
options = ['Sat', 'Sun', 'Mon']
df['day'].isin(options)
```

```python
# Get the restricted rows.
df[df['day'].isin(options)]
```
