# SQL Databases

```python
# Pandas can read in an entire table or sql query right into a dataframe.
from sqlalchemy import create_engine
import numpy as np
```

```python
temp_db = create_engine('sqlite:///:memory:')
```

```python
df = pd.DataFrame(data=np.random.randint(low=0, high=100, size=(4, 4)), columns=['a', 'b', 'c', 'd'])
```

```python
df
```

```python
# Create a new table in db from dataframe
df.to_sql(name='new_table', con=temp_db, index=False)
```

```python
# Cannot do two.
# Safeguard from overwriting data, can specify param to override it.
df.to_sql(name='new_table', con=temp_db)
```

```python
# Let's read!
new_df = pd.read_sql(sql='new_table', con=temp_db)
```

```python
new_df
```

```python
# Or pass in entire query!
pd.read_sql_query(sql='SELECT a, c FROM new_table;', con=temp_db) # Wow!
```
