# DataFrames

A group of pandas series objects that share the same index.
It is essentially like a DB table.
- The main pandas objects with work with.

```python
import numpy as np
import pandas as pd
```

```python
# Create a DataFrame

np.random.seed(101)
mydata = np.random.randint(0, 101, (4, 3))
```

```python
mydata
```

```python
# 4 indexes for 4 rows
myindex = ['CA', 'NY', 'AZ', 'TX']
mycolumns = ['Jan', 'Feb', 'Mar']
```

```python
# Only provide data
df = pd.DataFrame(mydata)
df
```

```python
df = pd.DataFrame(data=mydata, index=myindex)
df
```

```python
df = pd.DataFrame(data=mydata, index=myindex, columns=mycolumns)
df
```

```python
# Useful info on the DataFrame
df.info()
```

# Columns

```python
# Fetch one column series. Can also autocomplete.
df['total_bill']
```

```python
type(df['total_bill']) # It's a panda series!!
```

```python
# We need to pass in a "List" of column names not individual params.
df[['total_bill', 'tip']]
```

```python
# Create a new column. We can do operations
df['tip'] + df['total_bill']
```

```python
df['tip'] / df['total_bill']
```

```python
# Get tip percentage. Damn!!
100 * df['tip'] / df['total_bill']
```

```python
# Create new column.
# Ref as if it already exists.
df['tip_percentage'] = (100 * df['tip'] / df['total_bill'])

# We have added a new column!
df
```

```python
# We can use numpy functions for DataFrames!
df['price_per_person'] = np.round((100 * df['total_bill'] / df['size']), 2)
df
```

```python
# Remove columns
# Syntax for index or columns.
# axis = 0 drop drow, 1 = drop column
df.drop('tip_percentage', axis=1)
```

```python
# Any function calls on DataFrames do not ooccur in place
df
```

```python
# We can make the drop in place but this is better. inplace will be eventually deprecated.
df = df.drop('tip_percentage', axis=1)
```

# Remove and Insert new rows

```python
df.drop('Sun2959') # axis = 0 is row and that's what we want (it is default)
```

```python
df.head()
```

```python
df = df.drop('Sun2959') # Permenently dropped.
df.head()
```

```python
# Drop off integer location can not be done via drop,
# we could however just select the rows we want and replace the df like:
df = df.iloc[1:] # delete the first row 
```

```python
# Insert new row (must match dimensions)
one_row = df.iloc[0]
```

```python
one_row # Must be correct length and shape for us to append it
```

```python
# Duplicate a row
df = df.append(one_row)
df
```
