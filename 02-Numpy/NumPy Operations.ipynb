# NumPy Operations

```python
import numpy as np
```

```python
arr = np.arange(0, 10)
```

```python
arr
```

```python
# Arithmetic between arrays.
arr + 5 # element by element basis!
```

```python
arr
```

```python
arr - 2
```

```python
arr + arr # Must be same shape, adds each index in one to the other
```

```python
arr - arr
```

```python
arr
```

```python
# Numpy gives nan if 0 / 0
arr / arr
```

```python
1 / 0
```

```python
1 / arr
```

```python
# Division by 0 in numpy gives a warning but not an exception and puts nan in place instead.
```

```python
np.sqrt(arr)
```

```python
np.sin(arr)
```

```python
np.log(arr)
```

```python
# Perform summary stats on dataset
arr.sum() # Sum of all elements
```

```python
arr.mean()
```

```python
arr.max()
```

```python
arr.var() # Variance
```

```python
arr.std() # Standard deviation
```

```python
arr2d = np.arange(0, 25).reshape(5, 5)
```

```python
arr2d
```

```python
arr2d.sum()
```

```python
# Get sum of all the rows or all the columns?
arr2d.sum(axis=0) # Row = axis 0 (perform the operation across the rows)
```

```python
arr2d.sum(axis=1) # Row = axis 0 (perform the operation across the columns)
```
