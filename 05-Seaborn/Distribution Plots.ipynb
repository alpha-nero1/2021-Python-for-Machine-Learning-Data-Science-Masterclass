# Distribution Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize=(5, 8), dpi=200)
# Check out distribution of peoples salaries
# RUGPLOT!
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
# High level style changing command.
sns.set(style='darkgrid') # Gives a darker background with a white grid (white, light, lightgrid)

# HISTOGRAM
sns.displot(
    data=df,
    x='salary',
    # bins=100 Can edit binsize.
    # Matplotlib params...
    color='red',
    edgecolor='blue',
    linewidth=4,
    ls='--',
    kde=True, # Show Kernel Density Estimate
    rug=True
)
```

```python
# ALSO HISTOGRAM
sns.histplot(data=df, x='salary')
```

```python
# ONLY KDE
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
randomages = np.random.randint(0, 100, 200)
```

```python
randomages
```

```python
randomages = pd.DataFrame(randomages, columns=['age'])
```

```python
randomages
```

```python
sns.rugplot(data=randomages, x='age')
```

```python
sns.displot(data=randomages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(
    data=randomages, 
    x='age',
    clip=(0, 100), # Cut em off
    shade=True
)
```
