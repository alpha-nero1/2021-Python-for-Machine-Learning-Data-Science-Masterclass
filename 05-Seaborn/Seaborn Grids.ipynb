```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# CATPLOT
sns.catplot( # Uses sns.FacetGrid behind the scenes.
    data=df,
    x='gender',
    y='math score',
    kind='box',
    col='lunch', # Similar to hue but creates sub plots
    row='test preparation course'
)
```

```python
# PAIRPLOT
g = sns.PairGrid(
    df,
    hue='gender'
)

# Configure diagrams manually.
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.kdeplot)
# Need to add legend manually.
g= g.add_legend()
```
