# Categorical Plots

- Visualisation of a group by call.
- Show data for categories

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10, 4), dpi=200)
# COUNTPLOT
sns.countplot(
    data=df, 
    x='level of education',
    hue='division',
    palette='Set2'
)

# plt.ylim(90, 260) Careful of deceiving y-lims
```

```python
df['level of education'].value_counts()
```

```python
# BARPLOT
plt.figure(figsize=(10, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```
